マンカラのログの作成

```python
import random

#対戦ログ配列
A_win_log = []
B_win_log = []

#ゲームの終了条件
def end_game_flag(state):
  if state[0] == 0 and state[1] == 0:
    return "A"
  elif state[3] == 0 and state[4] == 0:
    return "B"
  return 0

for step in range(20):
  #盤面の初期化
  state = [3, 3, 0, 3, 3, 0]

  #プレイ中のプレイヤー (0 or 3)
  player = 0

  #1回の対戦ログ
  choice_log = []

  while True:
    #1ターンの処理
    #石を選ぶ
    choice_num = random.randint(0, 1)
    #print(choice_num)
    if state[choice_num+player] == 0:
      choice_num = (choice_num+1)%2 + player
    else:
      choice_num = choice_num + player
    #print(choice_num)

    #対戦ログに選んだ手を記入
    if choice_num == 3:
      choice_log.append("A")
    elif choice_num == 4:
      choice_log.append("B")
    else:
      choice_log.append(str(choice_num+1))

    #攻め継続判定
    next = 0
    print(choice_num + state[choice_num])
    if (choice_num + state[choice_num]) % 6 != 2 and (choice_num + state[choice_num]) % 6 != 5:
      next = 3
    else:
      print("攻め継続")

    #石を動かす
    select_hole = 0
    for stone in range(state[choice_num]+1):
      state[(choice_num+stone)%6] += 1
      select_hole+=1

    #選んだ穴の石の数をもとの数を引いたものにする
    state[choice_num] -= select_hole

    print("player"+str(player)+"が")
    print(str(choice_num)+"を選び")
    print(state)

    #ゲームの終了判定
    if end_game_flag(state) == "A":
      print("勝者 = A")
      if choice_log not in A_win_log:
        A_win_log.append(choice_log)
      break
    if end_game_flag(state) == "B":
      print("勝者 = B")
      if choice_log not in B_win_log:
        B_win_log.append(choice_log)
      break
    print("==")

    #次のプレイヤー
    player = (player+next)%6
```

```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

tokenizer = AutoTokenizer.from_pretrained("stabilityai/StableBeluga-13B", use_fast=False)
model = AutoModelForCausalLM.from_pretrained("stabilityai/StableBeluga-13B", load_in_8bit=True, low_cpu_mem_usage=True, device_map="auto")
system_prompt = "### System:\nYou are StableBeluga, an AI that follows instructions extremely well. Help as much as you can. Remember, be safe, and don't do anything illegal.\n\n"
```

```python
message = f"There is a turn-based game \
The first player can only choose between 1 and 2. \
The second player can only choose between A and B. \
1, 2, A and B are independent of each other and have no relation to each other. \
The result of the game is a log of the moves chosen in chronological order, indicating whether the winner is the first or the second player. \
Sample game results:. \
[first move of turn 1, second move of turn 1, first move of turn 2, second move of turn 2] = winner \
Match Result: \
{first_log} \
{second_log} \
Please answer step-by-step while thinking about the reason. \
Match Progress: \
[A2,B1] \
Which move should the first player choose next?"

prompt = f"{system_prompt}### User: {message}\n\n### Assistant:\n"
inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
output = model.generate(**inputs, do_sample=True, top_p=0.95, top_k=0, max_new_tokens=256)

print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
ans = tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
for a in ans.split("\n\n"):
  print(a)
```

```python
ai_choice = ans.split("\n\n")[2].split(":\n ")[1]
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

flan_tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-xl")
flan_model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-xl", device_map="auto")
```

```python
flan_input_text = ai_choice

flan_input_ids = flan_tokenizer(flan_input_text, return_tensors="pt").input_ids.to("cuda")

flan_outputs = flan_model.generate(flan_input_ids)
print(flan_tokenizer.decode(flan_outputs[0]))
```

```python
ans = flan_tokenizer.decode(flan_outputs[0])
```
